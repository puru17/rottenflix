```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Step 1: Load the image using OpenCV (JPG file)
image_path = "C:\\Users\\ktp17\\Desktop\\dev\\Rotten Flix ML Goated\\IMG_2706.jpg"  # Replace with the path to your image
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)  # Load in grayscale
plt.imshow(image, cmap='gray')
```

```python
# Hyperparameter ranges to test
kernel_sizes = [(3, 3), (5, 5), (7, 7), (9, 9), (11, 11)]
lower_thresholds = [30, 50, 70, 90, 110]
upper_thresholds = [100, 150, 200, 250, 300]
sigmas = [0, 1, 2, 3, 4]  # Add sigma values

# ... (rest of your code)

# Iterate over hyperparameter combinations
for kernel_size in kernel_sizes:
    for lower_threshold in lower_thresholds:
        for upper_threshold in upper_thresholds:
            for sigma in sigmas:
                # Step 2: Apply Gaussian blur
                blurred = cv2.GaussianBlur(image, kernel_size, sigma)

                # Step 3: Apply Canny edge detection
                edges = cv2.Canny(blurred, lower_threshold, upper_threshold)

                filename = f"hyp_testing\\edges_{kernel_size}_{lower_threshold}_{upper_threshold}_{sigma}.jpg"
                print(f"hyp_testing\\edges_{kernel_size}_{lower_threshold}_{upper_threshold}_{sigma}.jpg")
                cv2.imwrite(filename, edges)
```
